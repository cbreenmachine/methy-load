```r
library(Rtsne)
library(tidyverse)
library(wiscR)
```

```r
df <- read_csv("../../data/prin-comps/chr18.csv")
```

```r
tsne.fit <- df %>%
    select(starts_with("PC")) %>%
    scale() %>%
    Rtsne(perplexity = 32, pca = FALSE)

tsne.fit$Y %>%
    as.data.frame() %>%
    ggplot(aes(x = V1, y = V2)) +
    geom_point()
```
